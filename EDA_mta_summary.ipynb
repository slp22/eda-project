# Addressing Health Equity with Mobile COVID Testing

## **Abstract**

This preliminary analysis explored NYC subway entries during late-night hours using the [NYC subway turnstile data](web.mta.info/developers/turnstile.html). For stations with high late-night entries, the NYC health department could consider deploying mobile COVID testing units. In addition, providing services in low-income neighborhoods, whose residents may likely be essential workers, could fill a healthcare gap and protect people vulnerable to COVID because of shift work and unpaid sick leave. This analysis showed high entry traffic at various outer-borough stations, particularly in Queens. 

## **Design**
In the first year of the COVID pandemic, the New York borough of Queens experienced [a high burden of COVID disease and deaths](https://www1.nyc.gov/site/doh/covid/covid-19-data-neighborhoods.page). Zipcodes with high proportions of [low-income](https://censusreporter.org/profiles/86000US11368-11368/) residents were most affected. Essential works were especially vulnerable because of shift work and unpaid sick leave. This analysis uses income as a proxy for essential workers during the coronavirus pandemic.

The [New York City Department of Health and Mental Hygience](https://www1.nyc.gov/site/doh/about/about-doh.page) (NYC DOH) has [mobile testing units](https://www1.nyc.gov/site/coronavirus/get-tested/covid-19-testing.page) which often set up near subway stations. However, the testing units operate during peak hours (8 am - 8 pm), and late-night commuters may not have access to these services. The results of this analysis could provide NYC DOH with evidence to deploy units during late-night hours to stations in zipcodes with high numbers of essential workers. Serving people in NY who cannot access COVID services during regular business hours helps mitigate surges of COVID. Serving this population would also address health equity issues for shift workers who are often low-wage workers and have limited access to healthcare. Providing COVID monitoring and prevention services during late-night hours would benefit New Yorkers during their commute, despite irregular work hours. Furthermore, it would impact New York City's response to COVID surges by providing services to people they may otherwise not easily reach.

**Research Question:** Could the city health department address health equity by deploying mobile COVID testing units to subway stations during late-night hours to reach people who cannot access COVID services during regular hours?

## **Data**
This analysis uses the public [MTA Turnstile dataset](http://web.mta.info/developers/turnstile.html) dataset, reported every Saturday. The selected data explored data from April to June 2019, looking closely at subway entries during late-night hours (12 am - 4 am). The analysis first explores late-night data for all stations. Then follows a closer look at the *7* train stations through Queens. And finally, it examines traffic at Flushing-Main Street, a *7* train terminus.

MTA turnstile data for 2020 and 2021 is irregular because the MTA suspended 24-hour subway service from [May 2020](https://www.nytimes.com/2020/05/06/nyregion/nyc-subway-close-coronavirus.html) until [May 2021](https://www.nytimes.com/2021/05/17/nyregion/nyc-subway-full-service-24-hours.html). Therefore, the best approximation for late-night travel was the 2019 data used here. 

The [NYT Coronavirus by Zip Code](https://www.nytimes.com/interactive/2020/nyregion/new-york-city-coronavirus-cases.html) informed the areas of New York with the highest burden of COVID. And the [HealthNY Test Site Finder](https://coronavirus.health.ny.gov/find-test-site-near-you) provided information on mobile COVID testing units, including location and hours of operation. 

The analysis showed that select subway entrances have high foot traffic late at night. Based on the exploration of stations on the *7* train in Queens, the terminus, Flushing-Main Street, and stations close to Manhattan would be strategic places for mobile testing units. Based on this analysis, we recommend that NYC DOH assign mobile COVID testing units to subway stations with high late-night entries to provide testing services for essential workers.
